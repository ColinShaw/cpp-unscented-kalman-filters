## Visualization of NIS

The following two sections simply load the lidar and radar specific output files from the UKF evaluation routine 
and plot the NIS element along with the Chi-squared value for two and three degrees of freedom, respectively, for 
the 95% confidence. 

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename = '../output_files/out1l.txt'
chi_squared_value = 5.991

nis = []
chi_squared = []

with open(filename, 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    for row in reader:
        nis.append(float(row[8]))
        chi_squared.append(chi_squared_value)

x = range(len(nis))        
plt.plot(x, nis)
plt.plot(x, chi_squared)
plt.show()
```
